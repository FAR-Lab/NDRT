```python
import pandas as pd 
import numpy as np 
import base64
import os
```

```python
data = pd.read_csv("../data/sample_sim_data_jul30.csv", sep=';')
```

```python
VELOCITY='A escooter velocity'
POSITION='A escooter Pos'
ROTATION = 'A escooter Rot'
ACCELERATION = 'A escooter accel'
STEERING = 'A escooter steering'
```

```python
data
```

```python
data.columns
```

```python
# A scooter velocity is stored as a float array, base64 encoded 
# decode it 
data[VELOCITY] = data[VELOCITY].apply(lambda x: np.frombuffer(base64.b64decode(x), dtype=np.float32))
# parse out into x, y, and z components 
data['v_x'] = data[VELOCITY].apply(lambda x: x[0])
data['v_y'] = data[VELOCITY].apply(lambda x: x[1])
data['v_z'] = data[VELOCITY].apply(lambda x: x[2])
```

```python
# A scooter position is stored as a float array, base64 encoded
# decode it
data[POSITION] = data[POSITION].apply(lambda x: np.frombuffer(base64.b64decode(x), dtype=np.float32))
# parse out into x, y, and z components
data['x'] = data[POSITION].apply(lambda x: x[0])
data['y'] = data[POSITION].apply(lambda x: x[1])
data['z'] = data[POSITION].apply(lambda x: x[2])
```

```python
data[ROTATION] = data[ROTATION].apply(lambda x: np.frombuffer(base64.b64decode(x), dtype=np.float32))
# parse out into x, y, and z components
data['r_x'] = data[ROTATION].apply(lambda x: x[0])
data['r_y'] = data[ROTATION].apply(lambda x: x[1])
data['r_z'] = data[ROTATION].apply(lambda x: x[2])

```

```python
# rename acceleration and steering columns for convenience 
data.rename(columns={ACCELERATION: 'a', STEERING: 'steering'}, inplace=True)
```

```python
# drop original columns
data = data.drop(columns=[VELOCITY, POSITION, ROTATION])
```
