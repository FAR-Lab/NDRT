```python
import matplotlib.pyplot as plt 
import geopandas as gpd 
import pandas as pd 
import numpy as np 

from tqdm import tqdm 
from glob import glob 

import os 
import sys 
```

```python
# load parsed data 
parsed_sample = pd.read_csv("../processed/sample_sim_data_jul31_parsed.csv")
```

```python
parsed_sample = gpd.GeoDataFrame(parsed_sample, geometry=gpd.points_from_xy(parsed_sample.x, parsed_sample.z))
```

```python
parsed_sample['x'].describe([0.01, 0.025, 0.05, 0.1, 0.25, 0.5, 0.75, 0.9, 0.95, 0.975, 0.99]) 
```

```python
parsed_sample['y'].describe([0.01, 0.025, 0.05, 0.1, 0.25, 0.5, 0.75, 0.9, 0.95, 0.975, 0.99])
```

```python
parsed_sample['z'].describe([0.01, 0.025, 0.05, 0.1, 0.25, 0.5, 0.75, 0.9, 0.95, 0.975, 0.99]) 
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))

parsed_sample.plot(ax=ax, markersize=1, alpha=0.5, column='ScenarioTime', legend=True,
                   legend_kwds={'label': "Elapsed Scenario Time [seconds]", 'orientation': "horizontal", 'shrink': 0.5, 'pad': 0.01})

ax.set_axis_off()

ax.set_title("Trajectory of Jul 31 Sample Data")

os.makedirs("../figures", exist_ok=True)
plt.savefig("../figures/jul31_sample_data.png", dpi=300, bbox_inches='tight')
```
